```python
res = requests.get("https://raw.githubusercontent.com/python-visualization/folium/master/examples/data/world-countries.json")
```

```python
url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
country_shapes = f'{url}/world-countries.json'
```

```python
df = df.sort_values("Country name", ascending=True)
df
```

```python
# getting a list of country names from geojson
world_map_data = json.loads(res.text)

names = []
for index in range(len(world_map_data["features"])):
    names.append(world_map_data["features"][index]["properties"]["name"])

names.sort()
# names
```

```python
df.replace("Serbia", "Republic of Serbia", inplace = True)
df.replace("Tanzania", "United Republic of Tanzania", inplace = True)
df.replace("Congo (Kinshasa)", "Democratic Republic of the Congo", inplace = True)
df.replace("United States", "United States of America", inplace = True)
```

```python
m = folium.Map()
```

```python
choropleth = folium.Choropleth(geo_data = country_shapes, 
                               data = df, 
                               columns = ["Country name", "Perceptions of corruption"], 
                               fill_color = "RdPu", 
                               nan_fill_color = "white", 
                               key_on = "feature.properties.name",
                               legend_name = "Ladder Score",
                               smooth_factor=0).add_to(m)
style_function = "font-size: 15px; font-weight: bold"


choropleth.geojson.add_child(folium.features.GeoJsonTooltip(["name"], style=style_function, labels=False))
```
