```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
plt.style.use("seaborn")
```

```python
df = pd.read_csv("happiness-score/2020.csv")
df
```

```python
fig = plt.figure(figsize =(25,10))

sns.violinplot( x=df["Regional indicator"], y=df["Ladder score"])
plt.xlabel("Region")
plt.title("Ladder Score in Regional Areas", fontsize=30)
```

```python
df["Regional indicator"].value_counts()
```

```python
fig = plt.figure(figsize = (30,30))

plt.scatter(x=df["Logged GDP per capita"], y=df["Explained by: Social support"], s=df["Ladder score"]**4, c=df["Ladder score"]**2, cmap="Blues", alpha = .4, edgecolors="black", linewidth=1)
plt.xlabel("Logged GDP per Capita", fontsize=20)
plt.ylabel("Explained by: Social Support", fontsize=20)
plt.title("Correlation between Social Support, GDP per Capita, and Ladder Score", fontsize=30)
```
