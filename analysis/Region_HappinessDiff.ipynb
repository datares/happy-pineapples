```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df_15 = pd.read_csv("2015.csv")
df_16 = pd.read_csv("2016.csv")
df_17 = pd.read_csv("2017.csv")
```

```python
df_15.head()
```

```python
df_15["happiness_diff"] = df_15["Standard Error"] * 1.96 * 2
```

```python
df_15["happiness_diff"].describe()
```

```python
df_15_happDiffer_Region = df_15["happiness_diff"].groupby(df_15["Region"]).mean()
df_15_happDiffer_Region
```

```python
plt.plot(df_15["Happiness Rank"], df_15["happiness_diff"])
```

```python
df_16.head()
```

```python
df_16.loc[df_16["happiness_diff"] == max(df_16["happiness_diff"]), ]
```

```python
df_16["happiness_diff"] = df_16["Upper Confidence Interval"] - df_16["Lower Confidence Interval"]
```

```python
df_16_happDiffer_Region = df_16["happiness_diff"].groupby(df_16["Region"]).mean()
```

```python
df_17["happiness_diff"] = df_17["Whisker.high"] - df_17["Whisker.low"]
```

```python
plt.plot(df_17["Happiness.Rank"], df_17["happiness_diff"])
plt.show()
```

```python
country_region_dict = df_15[["Country", "Region"]].set_index('Country').T.to_dict('index')["Region"]
country_region_dict
```

```python
def region_reassign(row):
    if country_region_dict.has_key(row["Country"]):   
        return country_region_dict[row["Country"]]
    else:
        return "NotFound"
```

```python
df_17["Region"] = df_17.apply(region_reassign, axis = 1)
df_17.tail()
```

```python
df_17["Region"].value_counts()
```

```python
df_17.loc[(df_17["Region"] == "NotFound"),]
```

```python
df_17.at[32, "Country"] = "Taiwan"
df_17.at[70, "Country"] = "Hong Kong"
df_17.at[146, "Country"] = "Sudan"
```

```python
df_17["Region"] = df_17.apply(region_reassign, axis = 1)
df_17.tail()
```

```python
df_17["Region"].value_counts()
```

```python
df_17_happDiffer_Region = df_17["happiness_diff"].groupby(df_17["Region"]).mean()
df_17_happDiffer_Region
```

```python
df_17_happDiffer_Region = df_17_happDiffer_Region.drop(labels=['NotFound'])
```

```python
plt.figure(figsize=(12,5))
fig = sns.barplot(x = df_17_happDiffer_Region.index, y = df_17_happDiffer_Region)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,5))
fig = sns.barplot(x = df_15_happDiffer_Region.index, y = df_15_happDiffer_Region)
plt.xticks(rotation=45)
plt.show()
```

```python
N = 10
ind = np.arange(N)  # the x locations for the groups
width = 0.2       # the width of the bars

fig = plt.figure(figsize=(12,5))
ax = fig.add_subplot(111)
rects1 = ax.bar(ind, df_15_happDiffer_Region, width)

rects2 = ax.bar(ind+width, df_16_happDiffer_Region, width)

rects3 = ax.bar(ind+2*width, df_17_happDiffer_Region, width)

# add some
ax.set_xlabel('Region', size = 15)
ax.set_ylabel('1.96 * 2 * StandardError(Happiness_Score)', size = 13)
ax.set_title('Region vs Length of Confidence Interval in Happiness Score', size = 15)
ax.set_xticks(ind + width)
ax.set_xticklabels(df_15_happDiffer_Region.index, rotation=45)

ax.legend( (rects1[0], rects2[0], rects3[0]), ('2015', '2016', '2017') )

plt.show()
```
