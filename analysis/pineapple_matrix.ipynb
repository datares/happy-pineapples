```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
plt.style.use("seaborn")
```

```python
df = df.drop(["Ladder score", "Standard error of ladder score", "upperwhisker", "lowerwhisker", "Dystopia + residual", "Ladder score in Dystopia"], axis = 1)
```

```python
df.drop(df.iloc[:, 8:], inplace = True, axis = 1)
```
